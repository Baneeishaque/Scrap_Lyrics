```python
from bs4 import BeautifulSoup
from selenium import webdriver
import soupsieve as sv
```

```python
chromeWebDriver = webdriver.Chrome("C:/Programs/chromedriver.exe")
```

```python
def get_results_from_url(url,initialization=False) :
    chromeWebDriver.get(url)
    pageSource = chromeWebDriver.page_source
    beautifulSoapContent = BeautifulSoup(pageSource, "lxml")
    for lyricResult in sv.select('div.sec-lyric.clearfix', beautifulSoapContent):
#         lyricTitle=sv.select_one('p.lyric-meta-title',lyricResult)
#         print("Lyric Title : "+sv.select_one('a',lyricTitle).text)
#         lyricArtist=sv.select_one('p.lyric-meta-artists',lyricResult)
#         print("Lyric Artist : "+sv.select_one('a',lyricArtist).text)
        print(sv.select_one('a',sv.select_one('p.lyric-meta-title',lyricResult)).text 
              + " : " + sv.select_one('a',sv.select_one('p.lyric-meta-artists',lyricResult)).text)
#         lyricTitle=sv.select_one('a',sv.select_one('p.lyric-meta-title',lyricResult)).text
#         lyricArtist=sv.select_one('a',sv.select_one('p.lyric-meta-artists',lyricResult)).text
#         print(lyricTitle + " - " + lyricArtist)
    if initialization : 
        pageCount=sv.select_one('a:last-child',sv.select_one('div.pager',beautifulSoapContent)).text
        for i in range(2,int(pageCount)+1) :
            get_results_from_url("https://www.lyrics.com/serp.php?st=fuck+how+i+feel&p="+str(i))
```
